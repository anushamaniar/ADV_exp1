## Columns name with meanings:

Order_Date: The date the product was ordered.

Aging: The time from the day the product is ordered to the day it is delivered.

Customer_id: Unique id created for each customer.

Gender: Gender of customer.

Device_Type: The device the customer uses to actualize the transaction (Web/Mobile).

Customer_Login_Type: The type the customer logged in. Such as Member, Guest etc.

Product_Category: Product category

Product: Product

Sales: Total sales amount

Quantity: Unit amount of product

Discount: Percent discount rate

Profit: Profit

Shipping_cost: Shipping cost

Order_Priority: Order priority. Such as critical, high etc.

Payment_method: Payment method

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting Device Type Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Device_Type', palette='viridis')
plt.title('Distribution of Device Types Used by Customers')
plt.xlabel('Device Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting Customer Login Type Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Customer_Login_type', palette='muted')
plt.title('Distribution of Customer Login Types')
plt.xlabel('Login Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert 'Order_Date' to datetime
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True)

# Proceed with your analysis
df['Month'] = df['Order_Date'].dt.month_name()

# Now you can continue with your visualizations

# Sales by Month
df['Month'] = df['Order_Date'].dt.month_name()
plt.figure(figsize=(10, 6))
df.groupby('Month')['Sales'].sum().plot(kind='bar', color='skyblue')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()

# Sales by Day of the Week
df['DayOfWeek'] = df['Order_Date'].dt.day_name()
plt.figure(figsize=(10, 6))
df.groupby('DayOfWeek')['Sales'].sum().plot(kind='bar', color='coral')
plt.title('Total Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()

# Sales by Time of Day
df['Hour'] = pd.to_datetime(df['Time']).dt.hour
plt.figure(figsize=(10, 6))
df.groupby('Hour')['Sales'].sum().plot(kind='line', marker='o', color='green')
plt.title('Total Sales by Time of Day')
plt.xlabel('Hour')
plt.ylabel('Total Sales')
plt.show()
```

November had the highest sales, while February had the least

Tuesday had the highest sales, while Saturday had the least

- **Early Morning (0-6 AM)**: Sales are low, hitting the day's lowest point around 5 AM.
- **Morning (6-10 AM)**: Sales rapidly increase, peaking sharply at around 10 AM.
- **Midday to Evening (10 AM-8 PM)**: Sales remain high with some fluctuations, peaking again at noon and stabilizing at a high level until late evening.
- **Late Evening (8-11 PM)**: Sales gradually decline.

The highest sales occur in the late morning, while the early morning sees the least activity.

```python
# Calculating profit per unit
df['Profit_Per_Unit'] = df['Profit'] / df['Quantity']

# Plotting Profit Per Unit by Product
plt.figure(figsize=(12, 8))
df.groupby('Product')['Profit_Per_Unit'].mean().sort_values(ascending=False).head(10).plot(kind='bar', color='purple')
plt.title('Top 10 Products by Profit Per Unit')
plt.xlabel('Product')
plt.ylabel('Average Profit Per Unit')
plt.xticks(rotation=45)
plt.show()
```
